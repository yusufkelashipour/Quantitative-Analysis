```python
#First I will display all csv's from the relevant timeline to get a sense of what were working with
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 100)  # Show more rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)  # Wider display for columns
pd.set_option('display.colheader_justify', 'center')  # Center column headers

# Read the CSV files
df_1D = pd.read_csv('1D.csv')
df_60min = pd.read_csv('60min.csv')
df_5min = pd.read_csv('5min.csv')
df_1min = pd.read_csv('1min.csv')
df_10s = pd.read_csv('10s.csv')

for df in [df_1D, df_60min, df_5min, df_1min, df_10s]:
    df['ts_event'] = pd.to_datetime(df['ts_event'])

# Define the start and end dates for the week of May 27th, 2024
start_date = '2024-05-27'
end_date = '2024-05-31'

# Filter data for the specified date range
df_1D_filtered = df_1D[(df_1D['ts_event'] >= start_date) & (df_1D['ts_event'] <= end_date)]
df_60min_filtered = df_60min[(df_60min['ts_event'] >= start_date) & (df_60min['ts_event'] <= end_date)]
df_5min_filtered = df_5min[(df_5min['ts_event'] >= start_date) & (df_5min['ts_event'] <= end_date)]
df_1min_filtered = df_1min[(df_1min['ts_event'] >= start_date) & (df_1min['ts_event'] <= end_date)]
df_10s_filtered = df_10s[(df_10s['ts_event'] >= start_date) & (df_10s['ts_event'] <= end_date)]

print("1 Day Interval Data (2024-05-27 to 2024-05-31):")
display(df_1D_filtered)

print("\n60 Minute Interval Data (2024-05-27 to 2024-05-31:")
df_60min_head = df_60min_filtered.head(5)  
df_60min_tail = df_60min_filtered.tail(5)  
display(pd.concat([df_60min_head, df_60min_tail]))  

print("\n5 Minute Interval Data (2024-05-27 to 2024-05-31):")
display(df_5min_filtered)

print("\n1 Minute Interval Data (2024-05-27 to 2024-05-31):")
display(df_1min_filtered)

print("\n10 Second Interval Data (2024-05-27 to 2024-05-31):")
display(df_10s_filtered)
```

```python
#Next we need to understand what are important prices and ranges that should be analyzed.
#1. Significant Price Levels.
#We will anaylze these price levels using three different intervals to ensure consistency amongst them.
import matplotlib.pyplot as plt

# Extended range allows us to see if support and resistance levels are consistent
# and helps in identifying long-term trends or anomalies.
start_date = '2024-05-01'  
end_date = '2024-06-14'   

# Read and prepare data for the extended range
df_1D = pd.read_csv('1D.csv')
df_1D['ts_event'] = pd.to_datetime(df_1D['ts_event'])
df_1D_filtered = df_1D[(df_1D['ts_event'] >= start_date) & (df_1D['ts_event'] <= end_date)]

df_1min = pd.read_csv('1min.csv')
df_1min['ts_event'] = pd.to_datetime(df_1min['ts_event'])
df_1min_filtered = df_1min[(df_1min['ts_event'] >= start_date) & (df_1min['ts_event'] <= end_date)]

# Identify support and resistance levels for extended range
support_level_1D = df_1D_filtered['low'].min()
resistance_level_1D = df_1D_filtered['high'].max()

support_level_1min = df_1min_filtered['low'].min()
resistance_level_1min = df_1min_filtered['high'].max()

print(f"Support Level (1D): {support_level_1D}")
print(f"Resistance Level (1D): {resistance_level_1D}")
print(f"Support Level (1min): {support_level_1min}")
print(f"Resistance Level (1min): {resistance_level_1min}")

# Plotting
plt.figure(figsize=(14, 8))

# 1D Data
plt.plot(df_1D_filtered['ts_event'], df_1D_filtered['high'], label='1D High Prices', marker='o', linestyle='-', alpha=0.6)
plt.plot(df_1D_filtered['ts_event'], df_1D_filtered['low'], label='1D Low Prices', marker='o', linestyle='-', alpha=0.6)
plt.axhline(y=support_level_1D, color='g', linestyle='--', label='1D Support Level')
plt.axhline(y=resistance_level_1D, color='r', linestyle='--', label='1D Resistance Level')

# 1min Data
plt.plot(df_1min_filtered['ts_event'], df_1min_filtered['high'], label='1min High Prices', marker='o', linestyle=':', alpha=0.6)
plt.plot(df_1min_filtered['ts_event'], df_1min_filtered['low'], label='1min Low Prices', marker='o', linestyle=':', alpha=0.6)
plt.axhline(y=support_level_1min, color='g', linestyle=':', label='1min Support Level')
plt.axhline(y=resistance_level_1min, color='r', linestyle=':', label='1min Resistance Level')

plt.xlabel('Date and Time')
plt.ylabel('Price')
plt.title('Comparison of Support and Resistance Levels (Extended Timeline)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Volume Analysis

# Read 60-minute data
df_60min = pd.read_csv('60min.csv')
df_60min['ts_event'] = pd.to_datetime(df_60min['ts_event'])

df_60min_filtered = df_60min[(df_60min['ts_event'] >= '2024-05-01') & (df_60min['ts_event'] <= '2024-06-14')].copy()

# Calculate average volume for the filtered 60-minute data
average_volume_60min = df_60min_filtered['volume'].mean()

# Identify volume spikes (where volume is significantly higher than the average)
volume_spikes_60min = df_60min_filtered[df_60min_filtered['volume'] > 1.5 * average_volume_60min]

# Calculate average price (mean of high and low)
df_60min_filtered['average_price'] = (df_60min_filtered['high'] + df_60min_filtered['low']) / 2


fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(df_60min_filtered['ts_event'], df_60min_filtered['average_price'], label='Average Price', color='green', marker='o', alpha=0.7)
ax1.set_ylabel('Average Price', color='green')
ax1.tick_params(axis='y', labelcolor='black')

# Create a secondary y-axis for volume
ax2 = ax1.twinx()
ax2.bar(df_60min_filtered['ts_event'], df_60min_filtered['volume'], label='Volume', color='lightblue', alpha=0.5)
ax2.axhline(y=average_volume_60min, color='purple', linestyle='--', label='Average Volume')
ax2.set_ylabel('Volume', color='lightblue')
ax2.tick_params(axis='y', labelcolor='black')

ax1.set_xlabel('Date and Time')
plt.title('60-Minute Interval: Average Price and Volume Analysis (May 1st - June 14th, 2024)')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Read Data
df_daily = pd.read_csv('1D.csv')
df_daily['ts_event'] = pd.to_datetime(df_daily['ts_event'])

# Calculate daily returns
df_daily['returns'] = df_daily['close'].pct_change()

# Calculate daily volatility (standard deviation of returns)
daily_volatility = df_daily['returns'].std()

print(f"Daily Volatility: {daily_volatility}")

# Plotting daily volatility
plt.figure(figsize=(10, 6))
plt.plot(df_daily['ts_event'], df_daily['returns'], label='Daily Returns', color='blue')
plt.axhline(y=daily_volatility, color='red', linestyle='--', label='Daily Volatility')
plt.axhline(y=-daily_volatility, color='red', linestyle='--')
plt.title('Daily Volatility Analysis')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Ranking System
#Load your 5-minute interval data
df_5min = pd.read_csv('5min.csv')
df_5min['ts_event'] = pd.to_datetime(df_5min['ts_event'])

#Calculate Support Level
latest_support = df_5min['low'].min()

#Calculate Resistance Level
latest_resistance = df_5min['high'].max()

#Calculate Average Volume
average_volume = df_5min['volume'].mean()

#Calculate Volume Volatility
volume_volatility = df_5min['volume'].std()

#Current price and volume
current_price = df_5min['close'].iloc[-1]
current_volume = df_5min['volume'].iloc[-1]

#Score calculations
def calculate_support_score(current_price, support_level):
    return max(0, (support_level - current_price) / support_level * 100)  # Closer to support = higher score

def calculate_resistance_score(current_price, resistance_level):
    return max(0, (current_price - resistance_level) / resistance_level * 100)  # Closer to resistance = higher score

def calculate_volume_score(current_volume, average_volume):
    return min((current_volume / average_volume) * 100, 100)  

def calculate_volatility_score(volatility):
    return min(volatility * 10, 100)  

#Calculate scores
support_score = calculate_support_score(current_price, latest_support)
resistance_score = calculate_resistance_score(current_price, latest_resistance)
volume_score = calculate_volume_score(current_volume, average_volume)
volatility_score = calculate_volatility_score(volume_volatility)

weights = {
    'support': 0.20,
    'resistance': 0.20,
    'volume': 0.30,
    'volatility': 0.30
}

# Calculate overall ranking score using weighted average
overall_ranking_score = (
    (support_score * weights['support']) +
    (resistance_score * weights['resistance']) +
    (volume_score * weights['volume']) +
    (volatility_score * weights['volatility'])
)

print(f"Latest Support Level: {latest_support}")
print(f"Support Score: {support_score:.2f}")
print(f"Latest Resistance Level: {latest_resistance}")
print(f"Resistance Score: {resistance_score:.2f}")
print(f"Average Volume: {average_volume:.2f}")
print(f"Volume Score: {volume_score:.2f}")
print(f"Volume Volatility: {volume_volatility:.2f}")
print(f"Volatility Score: {volatility_score:.2f}")
print(f"Overall Ranking Score: {overall_ranking_score:.2f}")
```
